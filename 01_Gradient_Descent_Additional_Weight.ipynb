```python
def update_weights(m, b, w, X1, X2, Y, learning_rate):
    m_deriv = 0
    b_deriv = 0
    w_deriv = 0
    N = len(X)
    # Ableitung MSE
    for i in range(N): # pytorch - backward # gradient tape
        # Calculate partial derivatives
        # -x (innere Ableitung) * 2(y - (mx + b)) (채ussere Ableitung)
        m_deriv += -2*X1[i] * (Y[i] - (m*X1[i] + w*X2[i] + b))

        # -2(y - (mx + b))
        b_deriv += -2*(Y[i] - (m*X1[i] + w*X2[i] + b))

        w_deriv += -2*X2[i] * (Y[i] - (m*X1[i] + w*X2[i] + b))


    # We subtract because the derivatives point in direction of steepest ascent
    m -= (m_deriv / float(N)) * learning_rate # sp채ter - in pytorch step
    b -= (b_deriv / float(N)) * learning_rate
    w -= (w_deriv / float(N)) * learning_rate
    history_m.append(m)
    history_b.append(b)
    history_w.append(w)

    return m,b,w
```

```python
def train(epochs):
    m=0
    b=0
    w=0
    learning_rate=0.01
    for i in range(epochs):
        # abbruchskriterium - "early stopping"
        m,b,w = update_weights(m, b, w, X1, X2, Y, learning_rate)
    return m,b,w
```
