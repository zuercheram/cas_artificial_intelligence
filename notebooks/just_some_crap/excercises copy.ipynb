```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return 2 + np.sin(x) * np.cos(2*x-15)*2

def fy(x):
    return 2 * np.cos(x) * np.cos(2 * x - 15) - 4 * np.sin(x) * np.sin(2 * x - 15)

def calcCyclicLr(epoch, stepsize, lr, maxLr):
    cycle = math.floor(1 + epoch / (2 * stepsize))
    print(cycle)
    x = abs(epoch / stepsize - 2 * cycle + 1)
    return lr + (maxLr - lr) * max(0, (1-x))
    
x = np.linspace(0, 8, 200)
y = f(x)
```

```python
x_start = 7.5
dy_dx_old = fy(x_start)
dy_dx_new = 0

lr = 0.1
maxLr = 1
stepsize = 0.15
tolerance = 1e-1
index = 0
```

```python
plt.scatter(x_start, f(x_start))
while abs(dy_dx_new - dy_dx_old) > tolerance:
    dy_dx_old = dy_dx_new
    x_start = x_start - calcCyclicLr(index, stepsize, lr, maxLr) * dy_dx_old
    dy_dx_new = fy(x_start)
    #plt.scatter(x_start, f(x_start))
    index+=1
plt.scatter(x_start, f(x_start))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.show()
```
